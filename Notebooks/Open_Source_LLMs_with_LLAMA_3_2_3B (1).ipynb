
# Loading Llama 3.2 3B

````python
from transformers import AutoTokenizer
import transformers
import torch
````

````python
#specify the model you want to use
model = "meta-llama/Llama-3.2-3B-Instruct"
````

````python
#load the pretrained tokenizer
tokenizer = AutoTokenizer.from_pretrained(model)
````

````python
#make it ready for text generation
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
````

````python
messages = [{"role": "system", "content": "You are a helpful, respectful and honest assistant.Always answer as helpfully as possible, while being safe."},
    {"role": "user", "content": "draft a report on machine learning"}]
````

````python
outputs = pipeline(messages,max_new_tokens=256)
````
